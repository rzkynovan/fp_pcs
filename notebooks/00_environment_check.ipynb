## üóÇÔ∏è Check Dataset Files

```python
import glob

dataset_path = project_root / 'dataset'

print("="*60)
print("DATASET CHECK")
print("="*60)

if dataset_path.exists():
    # Find all EDF files
    edf_files = list(dataset_path.glob('*.edf'))
    
    # Separate healthy and MDD
    healthy_files = [f for f in edf_files if f.name.startswith('H ')]
    mdd_files = [f for f in edf_files if f.name.startswith('MDD ')]
    
    print(f"Total EDF files found: {len(edf_files)}")
    print(f"  - Healthy subjects: {len(healthy_files)}")
    print(f"  - MDD subjects: {len(mdd_files)}")
    
    if len(healthy_files) == 28 and len(mdd_files) == 30:
        print("\n‚úÖ Dataset complete! (28 Healthy + 30 MDD = 58 files)")
    else:
        print(f"\n‚ö†Ô∏è  Expected 28 Healthy + 30 MDD files, found {len(healthy_files)} + {len(mdd_files)}")
    
    # Show first 5 files
    print("\nFirst 5 files:")
    for i, file in enumerate(edf_files[:5], 1):
        print(f"  {i}. {file.name}")
else:
    print("‚ùå Dataset directory not found!")
    print("   Please create 'dataset/' folder and place your EDF files there.")
```

## üß™ Test Basic EDF Loading

```python
print("="*60)
print("EDF LOADING TEST")
print("="*60)

if len(edf_files) > 0:
    test_file = edf_files[0]
    print(f"Testing with: {test_file.name}\n")
    
    try:
        # Try with pyedflib
        import pyedflib
        
        f = pyedflib.EdfReader(str(test_file))
        n_channels = f.signals_in_file
        sample_rate = f.getSampleFrequency(0)
        duration = f.file_duration
        
        print("‚úÖ Successfully loaded EDF file!")
        print(f"   Channels: {n_channels}")
        print(f"   Sampling Rate: {sample_rate} Hz")
        print(f"   Duration: {duration:.2f} seconds")
        
        # Get channel labels
        print(f"\n   Channel labels:")
        for i in range(n_channels):
            label = f.getLabel(i)
            print(f"     {i+1}. {label}")
        
        f.close()
        
    except Exception as e:
        print(f"‚ùå Error loading EDF file: {e}")
else:
    print("‚ö†Ô∏è  No EDF files found in dataset directory")
```
